```python
labels = clicks["advertiser_id"].value_counts().index
sizes = clicks["advertiser_id"].value_counts().values
colors = ['yellowgreen', 'gold', 'lightskyblue', 'lightcoral', 'red', 'black', 'blue']
explode = (0, 0, 0, 0, 0, 0, 0)  # explode 1st slice
 
# Plot
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
autopct='%1.1f%%', shadow=True, startangle=140)
plt.axis('equal')
plt.show()

patches, texts = plt.pie(sizes, colors=colors, shadow=True, startangle=90)
plt.legend(patches, labels, loc="best")
plt.axis('equal')
plt.tight_layout()
plt.show()
```

```python
adv_id = clicks["advertiser_id"]
aux = adv_id.value_counts()
aux.plot(kind = 'bar')
plt.show()
```

```python
"""con estos graficos se peude ver como practicamente todos los advertiser_id son del tipo 3.Hay muy pocos de los otros en comparacion"""
```
