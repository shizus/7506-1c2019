```python
clicksAndInstalls.loc[:, "created_x"] = pd.to_datetime(clicksAndInstalls["created_x"], errors = "coerce")
clicksAndInstalls.loc[:, "created_y"] = pd.to_datetime(clicksAndInstalls["created_y"], errors = "coerce")
```

```python
"""Arraco con las fechas y horarios de los de los clicks"""
dates = clicksAndInstalls['created_x'].dt.day
p = dates.value_counts()
p.sort_index().plot(figsize = (12,8))
plt.title('Cantidad de clicks e installs en comun por ref_hash en las distintas fechas que se toco el click')
plt.xlabel('Dias')
plt.ylabel('Cantidad de clicks e installs')
plt.show()
```

```python
times = clicksAndInstalls['created_x'].dt.hour
q = times.value_counts()
q.sort_index().plot(figsize = (12,8))
plt.title('Cantidad de clicks e installs en comun por ref_hash en las distintas horas que se toco el click')
plt.xlabel('Horas')
plt.ylabel('Cantidad de clicks e installs')
plt.show()
```

```python
advertisers = clicksAndInstalls[['created_x','advertiser_id']].groupby(clicksAndInstalls['created_x'].dt.hour).agg({'advertiser_id':'value_counts'})
advertisers['advertiser_id'].unstack().plot.area(figsize = (12,8), title = 'Clicks por hora del día y cliente')
plt.xlabel('Horas')
plt.ylabel('Cantidad de clicks e installs')
plt.show()
```

```python
sources = clicksAndInstalls[['created_x','source_id']].groupby(clicksAndInstalls['created_x'].dt.hour).agg({'source_id':'value_counts'})
sources['source_id'].unstack().plot.area(figsize = (12,8), title = 'Clicks por hora del día y source_id')
plt.xlabel('Horas')
plt.ylabel('Cantidad de clicks e installs')
plt.show()
```
