```python
labels = clicks["advertiser_id"].value_counts().index
sizes = clicks["advertiser_id"].value_counts().values
colors = ['yellowgreen', 'gold', 'lightskyblue', 'lightcoral', 'red', 'black', 'blue']
explode = (0, 0, 0, 0, 0, 0, 0)  # explode 1st slice
 
# Plot
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
autopct='%1.1f%%', shadow=True, startangle=140)
plt.axis('equal')
plt.show()

patches, texts = plt.pie(sizes, colors=colors, shadow=True, startangle=90)
plt.legend(patches, labels, loc="best")
plt.axis('equal')
plt.tight_layout()
plt.show()
```

```python
adv_id = clicks["advertiser_id"]
aux = adv_id.value_counts()
aux.sort_index().plot(kind = 'bar', figsize = (10, 20))
plt.title('Cantidad de clicks por advertiser_ids')
plt.xlabel('advertiser_ids')
plt.ylabel('Cantidad de clicks')
plt.show()
```

```python
"""con estos graficos se peude ver como practicamente todos los advertiser_id son del tipo 3.Hay muy pocos de los otros en comparacion"""
```

```python
dates = clicks['created'].dt.day
p = dates.value_counts()
p.sort_index().plot()
plt.show()
```

```python
times = clicks['created'].dt.hour
q = times.value_counts()
q.sort_index().plot(figsize = (12,8))
plt.show()
```

```python
advertisers = clicks[['created','advertiser_id']].groupby(clicks['created'].dt.hour).agg({'advertiser_id':'value_counts'})
advertisers['advertiser_id'].unstack().plot.area(figsize = (12,8), title = 'Clicks por hora del día y cliente')
plt.show()
```

```python
sources = clicks[['created','source_id']].groupby(clicks['created'].dt.hour).agg({'source_id':'value_counts'})
sources['source_id'].unstack().plot.area(figsize = (12,8), title = 'Clicks por hora del día y source_id')
plt.show()
```

```python
specs = clicks[['created','specs_brand']].groupby(clicks['created'].dt.hour).agg({'specs_brand':'value_counts'})
specs['specs_brand'].unstack().plot.area(figsize = (12,8), title = 'Clicks por hora del día y specss_brand')
plt.show()
```
